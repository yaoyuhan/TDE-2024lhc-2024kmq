```python
import numpy as np
from copy import deepcopy
from scipy.interpolate import interp1d
from astropy.time import Time
```

```python
import matplotlib
import matplotlib.pyplot as plt
fs= 9
matplotlib.rcParams['font.size']=fs
ms = 6
matplotlib.rcParams['lines.markersize']=ms
```

```python
from load_specfiles import load_tde_specfiles, bin_spec
```

```python
def custom_ax(ax, xmax = 800, ymax = 8):
    ax.tick_params(which = 'major', length = 4, top=True, direction = "in", right = True)
    ax.tick_params(which = 'minor', length = 2, top=True, direction = "in", right = True)
    
    xmajor = 1000
    xminor = 100
    ymajor = 1
    yminor = 0.2
    ax.yaxis.set_major_locator(plt.MultipleLocator(ymajor))
    ax.yaxis.set_minor_locator(plt.MultipleLocator(yminor))
    ax.xaxis.set_major_locator(plt.MultipleLocator(xmajor))
    ax.xaxis.set_minor_locator(plt.MultipleLocator(xminor))
    
    ax.set_xlabel('Rest-frame Wavelength ('+r'$\rm \AA$'+')', fontsize = fs+1)
    ax.set_ylabel("Scaled "+r'$f_{\lambda}$'+"+offset", fontsize = fs+1)
    plt.xticks(fontsize= fs+1) 
    plt.yticks(fontsize= fs+1) 
```

```python
def plot_source():
    name = "ZTF24aaoxmyb"
    dt = load_tde_specfiles(names = [name], skip_sdss = False, skip_hst = True)
    
    subdt = dt[name]
    z = 0.2045
    spoff = subdt["off"]
    tref = 60424.40988
    
    keys = list(subdt.keys())
    keys = np.array(keys)
    ix = np.array([x[:2]=="sp" for x in keys])
    keys = keys[ix]
    nsp = len(keys)
    
    ymax = len(keys)*spoff 
    xmin = 2600
    ymin = -3
    xmax = 8800#10300/(1+z)
    lw = 0.5
    
    plt.figure(figsize = (4.5, 5.2))
    ax = plt.subplot(111)
    
    for i in range(nsp):
        mykey = keys[i]
        mysp = subdt[mykey]
        wave = mysp["wave_obs"]
        flux = mysp["flux0_obs"]
        inst = mysp["instrument"]
        if inst == "DESI":
            eflux = mysp["eflux0_obs"]
            mask = mysp["mask"]
            wave_desi = deepcopy(wave)
            model_desi = mysp["flux_model0"]
            wave_desi = wave_desi[mask==0]
            model_desi = model_desi[mask==0]
            #plt.plot(wave, flux)
            #plt.plot(wave, eflux)
            wave = wave[mask==0]
            flux = flux[mask==0]
            ix = (wave>4000)#&(wave<9305)
            wave = wave[ix]
            flux = flux[ix]
            
        mymjd = Time(mysp["date"]).mjd
        myphase = (mymjd - tref) / (1 + z)
        myphase = np.round(myphase, 0)
        wave = wave /(1+z)
        subix = (wave>5000)&(wave<6000)
        mynorm = np.median(flux[subix])/1.4
        flux /= mynorm
        if inst == "DESI":
            wave_desi = wave_desi /(1+z)
            model_desi /= mynorm
        offset = (nsp-i-1) * spoff -0.5
        if i<2:
            offset -= 10.7
        if i<3:
            offset += 0.5
        
        if "binning" in mysp:
            binning = mysp["binning"]
        else:
            binning = 3
        ww, ff = bin_spec(wave, flux, binning = binning)
        color = "goldenrod"
        if inst=="DESI":
            color = "k"
        elif inst == "SDSS":
            color = "k"
        elif inst=="LRIS":
            color = "darkcyan"
        elif inst == "Kast":
            color = "darkcyan"
        
        elif inst == "FLOYDS-N":
            color = "darkcyan"
        ax.plot(wave, flux+offset, linewidth = 1, color = color, alpha = 0.05)
        ax.plot(ww, ff+offset, linewidth = lw, color = color, alpha = 1)
        
        myfunc = interp1d(ww, ff+offset)
        xpos = 7200
        ypos = myfunc(xpos)+spoff*0.3
        if i<2:
            ypos += 0.1
        inst_ = inst
        if myphase < -100:
            mytext = "Archival "+inst_
        elif myphase<0:
            mytext = "$-$%d d, "%(-1*myphase)+inst_
        else:
            mytext = "$+$%d d, "%(myphase)+inst_
        ax.text(xpos, ypos, mytext, fontsize = fs)
        
     
    custom_ax(ax, xmax, ymax)
    ax.text(2800, 9.5, "AT2024lhc", weight = "bold", fontsize = fs+1)
    
    ax.set_ylim(ymin, ymax)
    ax.set_xlim(xmin, xmax)
    ax.set_yticklabels([])
    plt.tight_layout(rect = (-0.02, -0.03, 1.03, 1.01))
    figname = "../figs/opt_spec_24lhc.pdf"
    
    plt.savefig(figname)
    plt.show()
    
```

```python
plot_source()
```
